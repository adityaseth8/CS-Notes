# Dynamic Programming Introduction
 - Technique where problem broken down into sub-problems, store results of subproblems
 - Use whenever we see a recursive solutions which requires repeated calculation of values
 - https://www.youtube.com/watch?v=mBNrRy2_hVs


```python
# Recursive Method for Fibonacci Numbers
def fibonacci(n):
    if n <= 1:
        return n
    # if n == 0:
    #     return 0
    # if n == 1:
    #     return 1
    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(6) # 0 + 1 + 1 + 2 + 3 + 5,,, 5+3 = 8
```

Issue with Recursive Method: repeated calculations


```python
# Geeks for Geek Dynamic Programming Method: Stores every value in memory
def fibonacci(n):
    arr = [0, 1]
    for i in range(2, n+1):
        arr.append(arr[i-1] + arr[i-2])
    return arr[n]

fibonacci(255)
```

```python
# Optimized Dynamic Programming Approach: Storing only prev 2 terms in memory

# Fibonacci Numbers using Bottom Up Pattern

def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    prevPrevNum = 0 #val, index
    prevNum = 1

    for i in range(2, n+1):
        temp = prevPrevNum + prevNum # 7
        prevPrevNum = prevNum # 4
        prevNum = temp  # 7
    return prevNum

        # 3 4 --> 4 7
    

fibonacci(6)
```
