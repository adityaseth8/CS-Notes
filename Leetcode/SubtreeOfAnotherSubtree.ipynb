# Subtree of Another Subtree

### LC Easy

Given the roots of two binary trees root and subRoot, return true if there is a subtree of root with the same structure and node values of subRoot and false otherwise.

A subtree of a binary tree tree is a tree that consists of a node in tree and all of this node's descendants. The tree tree could also be considered as a subtree of itself.

The number of nodes in the root tree is in the range [1, 2000].  
The number of nodes in the subRoot tree is in the range [1, 1000].  

```python
# Initial Solution

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isSubtree(self, root, subRoot):
        # initial sol: iterate through root until subRoot root node is found. 
        # from there do same checks as prev problem (Same tree) implementing prev problem neetcode recursive dfs from that step onwards
        
        def isSameTree(root, subRoot):
            # Base cases:
            if not root and not subRoot: # both null
                return True
            if not root or not subRoot or root.val != subRoot.val: # one null, other isnt
                return False
            return (isSameTree(root.left, subRoot.left) and isSameTree(root.right, subRoot.right)) # both non null
        
        subTreeBool = False
        frontierQueue = [root]
        while (frontierQueue and subTreeBool == False):
            node = frontierQueue.pop(0)
            if node.val == subRoot.val:
                subTreeBool = isSameTree(node, subRoot)
            if (node.right):
                frontierQueue.append(node.right)
            if (node.left):
                frontierQueue.append(node.left)
        return subTreeBool



        """
        :type root: TreeNode
        :type subRoot: TreeNode
        :rtype: bool
        """
```

```python
# Online Solution

# Time Complexity : O(Tree * SubTree) because every root in Tree could be start of subtree

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isSubtree(self, root, subRoot):
        if not root and not subRoot: return True
        if not root or not subRoot : return False
        if root.val == subRoot.val and self.sameTree(root, subRoot):
            return True
        else:
            return self.isSubtree(root.left, subRoot) or self.isSubtree(root.right, subRoot)
    
    def sameTree(self,p,q):
        if not p and not q:
            return True
        if not q or not p :
            return False
        return p.val == q.val and self.sameTree(p.left,q.left) and self.sameTree(p.right,q.right)
```
