# Diameter of Binary Tree

### LC Easy

Given the root of a binary tree, return the length of the diameter of the tree.

The diameter of a binary tree is the length of the longest path between any two nodes in a tree. This path may or may not pass through the root.

The length of a path between two nodes is represented by the number of edges between them.

Example 1:  
  
Input: root = [1,2,3,4,5]  
Output: 3  
Explanation: 3 is the length of the path [4,2,1,3] or [5,2,1,3].  

```python
"""
Initial Solution: Sorta brute force Method. Using BFS to iterate through every node in the tree, and
calculating the maximum diameter at that node. Diameter calculated by taking max height of L and R subtrees,
and adding 2. 
"""

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def findHeight(self, node):
        if not node:
            return 0
        lHeight = self.findHeight(node.left) + 1
        rHeight = self.findHeight(node.right) + 1
        return max(lHeight, rHeight)
    def diameterOfBinaryTree(self, root):
        # iteratively go through each node in the tree using BFS, and calulate max diameter
        maxDiameterOfTree = 0
        if not root:
            return 0

        frontierQueue = [root]

        while (frontierQueue):
            lHeight, rHeight = 0, 0
            node = frontierQueue.pop(0)
            # find max diameter from that node in the tree
            
            lHeight = self.findHeight(node.left)
            rHeight = self.findHeight(node.right)
            currDiameter = lHeight + rHeight
            maxDiameterOfTree = max(currDiameter, maxDiameterOfTree)

            if node.left:
                frontierQueue.append(node.left)
            if node.right:
                frontierQueue.append(node.right)
        return maxDiameterOfTree        
```
