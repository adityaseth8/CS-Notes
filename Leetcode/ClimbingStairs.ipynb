# Climbing Stairs

### LC Easy

You are climbing a staircase. It takes n steps to reach the top.  

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?  
 
Input: n = 3  
Output: 3  
Explanation: There are three ways to climb to the top.  
1. 1 step + 1 step + 1 step  
2. 1 step + 2 steps  
3. 2 steps + 1 step  

Used Decision Tree to visualize




```python
class Solution(object):
    def climbStairs(self, n):
        # dynamic programming
        # to prevent redundancy
        
        # array where index is n, and val is number of ways
        arr = [0, 1, 2]
        for i in range(3, n+1):
            arr.append(arr[i-1] + arr[i-2])
        return arr[n]

        """
        :type n: int
        :rtype: int
        """
```

```python
class Solution(object):
    def climbStairs(self, n):

        # memory efficient: store only prev 2 terms
        
        if n <= 2:
            return n
        
        prevTerm = 2
        prevPrevTerm = 1

        for i in range(3, n+1):
            temp = prevTerm + prevPrevTerm
            prevPrevTerm = prevTerm 
            prevTerm = temp
        return prevTerm
```

Neetcode Solution: Stqrting from bottom (n steps), solve base case, go up until hit 0

Base case: Starting at n (last step), there is only 1 way we can land here
Now at n-1: Only 1 way from n-1 to get to goal n.

Now at n-2: How many ways to get to goal n. Depends on 2 subproblems (from n-1 and from n). You just add those values together.



```python
class Solution(object):
    def climbStairs(self, n):
        one, two = 1, 1

        for i in range(n-1):
            temp = one
            one = one + two
            two = temp
        return one
```
